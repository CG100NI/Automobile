```python
# Data Acquistion
# Data wrangling
import pandas as pd
import numpy as np
dataset_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
automobile_data = pd.read_csv(dataset_url,header = None)

#printing the first 5 info of the dataset
automobile_data.head(5)
```

```python
# create the header list 
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location",
           "wheel-base","length","width","height","curnb-weight","engine-type","num-of-cylinders","engine_size","fuel-system","bore",
          "stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
#place header in the dataset 
automobile_data.columns = headers
automobile_data.head(5)
```

```python
#drop the price from the dataset
automobile_data.dropna(subset=["price"],axis=0)
```

```python
#steps for working with missing data 
#1.indentify the missing data 
#2.deal with the missing data 
#3.correct the data format
# import numpy also

#indentify the missing data 
#Convert ? to NAN 
automobile_data.replace("?",np.nan,inplace= True)
automobile_data.head(5)
```

```python
#evaluating missing data 
#.isnull(), .notnull()

missing_data = automobile_data.isnull()
missing_data.head(5)
```

```python
#count missing value in each columns

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#deal with the missing values

#1.drop the whole row
#2. Drop the whole columns
# Replace the missing values
# mean, frequency or else sometime with other functions


#Replace by mean
#Replace NAN by mean value

norm_loss = automobile_data["normalized-losses"].astype("float").mean(axis=0)
print("average of normalized-losses:",norm_loss)
```

```python
automobile_data["normalized-losses"].replace(np.nan,norm_loss,inplace=True)
```

```python
# to see the data whether its replaced or not 
automobile_data.head(5)
```

```python
automobile_data.info()
```

```python
#convert the data types to the proper format 
automobile_data[['normalized-losses']] =automobile_data[['normalized-losses']].astype("int")
```

```python
# Find the mean value in Bore

bore = automobile_data["bore"].astype("float").mean(axis=0)
print("average of bore:", bore)
```

```python
# Replace missing value in Bore
automobile_data["bore"].replace(np.nan,bore,inplace=True)
```

```python
#convert the data types to the proper format
automobile_data[['bore']] = automobile_data[['bore']].astype('float')
```

```python
# Find the mean value

stk = automobile_data["stroke"].astype("float").mean(axis=0)
print("average of stroke:",stk)
```

```python
# Replacing missing value in stroke
automobile_data["stroke"].replace(np.nan,stk,inplace=True)
```

```python
#convert data type in proper format
automobile_data[['stroke']] = automobile_data[['stroke']].astype('float')
```

```python
# Find the mean value

hrpower = automobile_data["horsepower"].astype("float").mean(axis=0)
print("average of horsepower:", hrpower)
```

```python
# Replacing missing value in horsepower
automobile_data["horsepower"].replace(np.nan,hrpower,inplace=True)
```

```python
#convert the data type in proper format
automobile_data[['horsepower']] = automobile_data[['horsepower']].astype('float')
```

```python
# Find the mean value

peakrpm = automobile_data["peak-rpm"].astype("float").mean(axis=0)
print("average of peakrpm:",peakrpm)
```

```python
# Replacing missing value in peak-rpm
automobile_data["peak-rpm"].replace(np.nan,peakrpm,inplace=True)
```

```python
#convert the data type in proper format
automobile_data[['peak-rpm']] = automobile_data[['peak-rpm']].astype('float')
```

```python
#create group name 
import numpy as np

bins = np.linspace(min(automobile_data["horsepower"]),max(automobile_data["horsepower"]),4)
```

```python
Group_name = ['low','Med','High']
```

```python
automobile_data['horsepower_bin']= pd.cut(automobile_data['horsepower'],bins,labels=Group_name,include_lowest=True)
automobile_data[['horsepower','horsepower_bin']].head(20)
```
